```python
import os
os.chdir(r"C:\Users\RexDe\Documents\Python\critical-minerals-canada")

import pandas as pd
import geopandas as gpd

gdf = gpd.read_parquet(r"data\01_raw.parquet")
gdf = gdf[gdf["MAT"].isin(["Stream Sediment and Water", "Stream Sediment"])]

li_cols = [c for c in gdf.columns if "Li" in c or "_li_" in c.lower()]
print("lithium columns and non-null counts:")
for col in li_cols:
    n = gdf[col].notna().sum()
    pct = n / len(gdf) * 100
    print(f"  {col:<25} {n:>6,}  ({pct:.1f}%)")
```
